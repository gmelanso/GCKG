```python
import pandas as pd

df = pd.read_csv("../entities/csv/ContractClaim.csv", header=0)
grouped = df.groupby(['vendor'])
n_claims= grouped.size().sort_values(ascending=False).iloc[0:5]
total_expense= grouped['expenseTotal'].sum().reindex(n_claims.index)

pd.DataFrame(
    {
        "n_claims": n_claims,
        "total_expense": total_expense
    }
)
```

```python
grouped= df.groupby(['vendor', 'reportedBy'])
total = grouped['expenseTotal'].sum().sort_values(ascending=False)

pd.DataFrame(
    {
        "total": total
    }
).iloc[0:10]
```
